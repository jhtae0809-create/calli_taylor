```python
import os
from twilio.rest import Client

# Twilio Setting
ACCOUNT_SID = '_'
AUTH_TOKEN = '_'
TWILIO_NUMBER = '_'

client = Client(ACCOUNT_SID, AUTH_TOKEN)

# Contact list
contacts = [
    {"name": "John", "company": "State College Plumbing", "phone": "+181480000001"},
    {"name": "John", "company": "State College Electrics", "phone": "+18140000000"}
]

def get_message(name, company, survey_url):
    msg = (f"Hi, sorry for the random text! I'm Hyuntae, a student at PSU. "
           f"I reached out via SMS because I truly value the expertise of local pros. "
           )
    return msg

def send_sms():
    short_url = "---"

    print(f"--- 발송 시작 (총 {len(contacts)}건) ---")

    for person in contacts:
        body_text = get_message(person['name'], person['company'], short_url)

        if len(body_text) > 160:
            print(f"ℹ️ 알림: {person['company']} 메시지가 160자를 초과함 ({len(body_text)}자) - 2세그먼트 발송")

        try:
            message = client.messages.create(
                body=body_text,
                from_=TWILIO_NUMBER,
                to=person['phone']
            )
            print(f"✅ 전송 성공: {person['company']} ({message.sid})")
        except Exception as e:
            print(f"❌ 전송 실패: {person['company']} - {str(e)}")

if __name__ == "__main__":
    send_sms()
```
